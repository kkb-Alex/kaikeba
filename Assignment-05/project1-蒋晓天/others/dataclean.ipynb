对所有wiki data都进行数据清洗的工作

```python
import opencc
import re

cc = opencc.OpenCC('t2s')
html_str = '(<doc.*>)|(</doc>)'
punctuation = '！？｡。，＂＃＄％＆＇（）＊＋－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃《》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘\'‛“”„‟…‧﹏' \
            + '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~·'

def wiki_data_clean(s):
    new = ''
    for line in s.split('\n'):
        if line:
        # t2s
            line = cc.convert(line)
    
            # html -> space
            line = re.sub(html_str, ' ', line)
            
            # specical characters -> space
            line = re.sub('[{}]'.format(punctuation), ' ', line)
            
            new = new + ' ' + line
    return new
```

>wiki的语料结构性很明显，每篇文章都由<doc\><\\doc\>包裹而成，因此这里将每一篇文章为一个段落来整理

```python
import opencc
import re
import jieba

cc = opencc.OpenCC('t2s')
html_str = '(<doc.*>)|(</doc>)'
punctuation = '！？｡。，＂＃＄％＆＇（）＊＋－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃《》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘\'‛“”„‟…‧﹏' \
            + '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~·'
    
def parse_raw_wikidata(read_file):
    with open(read_file, 'r') as r_file:
        write_file = read_file + '_parsed'
        with open(write_file, 'w') as w_file:
            # file is too large, read line by line
            line = r_file.readline()
            
            # repeat until all file is read
            while line:
                if line.strip('\n'):
                    # traditional to simple
                    line = cc.convert(line)
                    # html -> space
                    line = re.sub(html_str, ' ', line)
                    # specical characters -> space
                    line = re.sub('[{}]'.format(punctuation+'\n'), ' ', line)
                
                    # divide line
                    words = jieba.cut(line)
                
                    article = ' '.join([word.strip() for word in words])
                    if article.strip():
                        w_file.write(article+'\n')
                line = r_file.readline()
                
```

```python
files = ['wiki_00', 'wiki_01', 'wiki_02']
for file in files:
    parse_raw_wikidata(file)
```

##### 对汉语新闻语料库进行清洗

```python
import numpy as np
import pandas as pd
```

```python
news = pd.read_csv('sqlResult_1558435.csv', encoding='gb18030')
```

我们只需要content和title两列的数据

```python
news_article = news.loc[:, ['content', 'title']]
```

```python
news_article.head()
```

```python
news_article.shape
```

```python
for i in range(100):
    title, content = news_article.iloc[i, 1], news_article.iloc[i, 0]
    for s in [title, content]:
        # specical characters -> space
        s = re.sub('[{}]'.format(punctuation+'\n'), ' ', s)
        # divide line
        words = jieba.cut(s)
                
        article = ' '.join([word.strip() for word in words])
        print(article)
```

```python
with open('news_parsed', 'w') as news_w:
    for i in range(news_article.shape[0]):
        title, content = news_article.iloc[i, 1], news_article.iloc[i, 0]
        for s in [str(title), str(content)]:
            # specical characters -> space
            s = re.sub('[{}]'.format(punctuation+'\n'), ' ', s)
            # divide line
            words = jieba.cut(s)
                
            article = ' '.join([word.strip() for word in words])
            if article.strip():
                news_w.write(article+'\n')
```
