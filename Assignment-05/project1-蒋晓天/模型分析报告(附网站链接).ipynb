#### 模型优点

>1. 能按照原文的顺序返回相关度最高的5个句子，保证了摘要连贯性
2. ~~以我的语文水平来看，~~摘要总体的准确性不错
3. 摘要完美的保留了原文标点符号
4. 在长新闻中的表现很不错

>总体来说，模型效果不错，和百度的[AI开放平台](https://ai.baidu.com/tech/nlp_apply/news_summary?)相比，同一篇文章的摘要相似度挺高。说明建立的模型接近大厂水平(事实上我在12月5号才知道有这样一个开放平台，不然的话我可以把平台的结果作为benchmark来调参数，也不用靠自己的拙劣语文水平来判断模型效果了)

#### 模型缺点

模型的缺点还是不少的。或者说，在自动摘要这个命题下，该模型表现一般，应该要做的更好<br>
下面罗列一下缺点以及应该补足的地方<br><br>
**缺点**
>1. 这个项目其实不是传统意义上的摘要，只是挑出和全文以及标题语意接近的若干句子组成短文章，没有真正做到精简概要的摘要。
2. 没法在保持文章顺畅性的前提下把句子也进行缩短。比如说一个长句子虽然和文章本身相关性很高，但是长句中有很多连接词，或者重复表达同一个意思，这个模型没法改变句子内部的结果从而让句子更加简洁
3. 如果一篇文章重复的提出相同的主题，摘要很可能会是几句语意很接近的句子组成，看起来很冗余。比如说一篇新闻讲台风，然后3个段落都提高了台风非常非常强，这样的话模型选出来摘要可能是：台风很强，破坏力惊人；台风非常强，造成XXX损失；台风很强，因此大家出门注意安全。其实完全可以用一个句子来概括3个句子，但是由于这3个句子都和文章关系较大，模型会同时选择这3个句子。
4. 如果新闻很短，那么摘要的效果就会很差(一部分是因为我固定了摘要的句子数量n，其实也可以作为一个输入参数让用户来填，或者根据文章的句子总数来得到摘要的句子数量，这方面是可以补足的)，因为模型是以句子为最小单位，文章越长，效果越好
5. 模型没法很好的处理这种语句---->“今天天气真好啊！”路人B说道。模型会把说话的人和说的话分为2个句子，然后由于相关性，这两个句子很可能不会同时出现在摘要中，这样的话摘要的连贯性就会变得特别差。但是如果要把说话的人和说的话绑定在一起，我暂时无法做到，因为说话的人可能在句子前，可能在句子后，也可能在2个句子之间，可能文章里不会用说这个字，可能是同义词，也可能是“想”，总之如果要绑定在一起，我只能用if-else来做，而且每次遇到新的情况都必须更新代码，成本很高。

**补足**
>1. 代码通篇没有捕捉error的地方，默认输入为string，在遇到其他输入的情况下程序会停。其实是需要对输入进行严格的控制的
2. flask web没有做error.html。一旦出现httperror，或者inputerror，用户需要手动输入主页网站进行跳转。理想中的情况是出现error了，跳转到error.html，然后用户可以点击error页面的某个地方返回到主页。
3. web的设计不到位。其实一开始我希望的是设计成翻译网站那样，一边输入数据，然后点击按钮后会在另外一边出现摘要。但由于之前没接触过任何前端编写，没法在短时间内实现这样的设计，最后只能是一个页面收集数据，然后存入字典，跳转到另一个页面，显示字典里的数据。

>需要补足的内容会在课后学习了error相关模块以及flask后进行修改
